# advance join operations

```python
df1=pd.DataFrame({'employers':["Ali","Veli","Ayşe","Fatma"],
                 'group':["muhasebe","Mühendislik","Mühendislik","Ik"]})
df1
```

```python
df2=pd.DataFrame({'employers':["Ali","Veli","Ayşe","Fatma"],
                 'start_date':[2010,2009,2014,2019]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on="employers")
```

```python
# many to one
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df4=pd.DataFrame({'group':["muhasebe","Mühendislik","Ik"],
                 'manager':["Caner","Mustafa","Berkcan"]})
df4
```

```python
pd.merge(df3,df4)
```
